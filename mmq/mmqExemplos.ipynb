```python
import numpy as np

n = int(input())
arrayX = [float(input()) for x in range(n)]
arrayY = [float(input()) for x in range(n)]
x = np.array(arrayX).reshape((n,1))
y = np.array(arrayY).reshape((n,1))

X = np.concatenate((np.ones_like(x), x**1), 1)
XtX = np.transpose(X)@X
XtY = np.transpose(X)@y
a = np.linalg.solve(XtX, XtY)
VE = X@a - y
EQ = np.sum(VE**2)

soma = np.sum(a) + EQ
print(f'{soma:.4f}')
```

```python
import numpy as np

n = int(input())
arrayX = [float(input()) for x in range(n)]
arrayY = [float(input()) for x in range(n)]
x = np.array(arrayX).reshape((n,1))
y = np.array(arrayY).reshape((n,1))

Ht = np.log(y/x)
X = np.concatenate((np.ones_like(x), x), 1)
c = np.copy(Ht)
XTX = np.transpose(X)@X
XTc = np.transpose(X)@c

a = np.linalg.solve(XTX, XTc)
A = (np.exp(a[0]),-a[1])

VE = c - X@a
EQ = np.sum(VE**2)

soma = np.sum(A) + EQ
print(f'{soma:.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([
    [3.4],
    [5.6],
    [9.9],
    [1.3]
])

y = np.array([
    [2.3],
    [1.4],
    [1.1],
    [4.6]
])

X = np.concatenate((np.ones_like(x), x**1), 1)
print(X, '\n')

XtX = np.transpose(X)@X
XtY = np.transpose(X)@y
a = np.linalg.solve(XtX, XtY)
print(a, '\n')

VE = X @ a - y
EQ = np.sum(VE**2)

print(f'EQ: {EQ:.4f}\n')

xg = np.linspace(0, 10, 100)
yg = a[0] + a[1] * xg

plt.plot(xg, yg)
plt.plot(x, y, "o")
plt.show()
```
