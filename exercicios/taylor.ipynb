# Séries de Taylor

Aproximações de uma função.

## Importações

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## Função

```python
def f(x):
  return np.log(x ** 2 + 1)
```

## Derivadas

```python
def f1(x):
  return 2 * x / (x ** 2 + 1)

def f2(x):
  return 2 * (1 - x ** 2) / ((x ** 2 + 1) ** 2)

def f3(x):
  return 4 * x * (x ** 2 - 3) / ((x ** 2 + 1) ** 3)
```

## Questão 1

Esboce um gráfico da função f(x) nas proximidades de x0 = -0.51.

```python
x = np.linspace(-1, 2, 100)
y = f(x)
plt.plot(x, y, "green")

plt.title("Gráfico da função f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

## Questão 2

Escreva uma função para calcular o polinômio de Taylor de segundo grau P2(x) para a função f(x) em torno de x0 = -0.51 e x = -1.37.

```python
def p2(x, x0):
  return f(x0) + (f1(x0) * (x - x0)) + ((f2(x0) / 2) * ((x - x0) ** 2))

print(p2(-1.37, -0.51))
```

## Questão 3

Calcule o erro absoluto cometido na aproximação entre f(x) e P2(x).

```python
erro_absoluto = np.abs(f(-1.37) - p2(-1.37, -0.51))
print(erro_absoluto)
```

## Questão 4

Plote um gráfico com a função f(x) e o polinômio P2(x) na região de interesse da aproximação.

```python
plt.plot(x, y, "green")
plt.plot(x, p2(x, -0.51), "red")

plt.title("Gráfico da função f(x) e do polinômio P2(x)")
plt.legend(["F(x)", "P2(x)"])
plt.grid()
plt.show()
```

## Questão 5

Com valores fixos de x0 = -0.51 e x = -1.37, escreva uma função para retornar um valor para o resto R2(t) e plotar os resultados em um gráfico.

```python
def r2(t, x0, x):
  return (f3(t) / 6) * ((x - x0) ** 3)

t = np.linspace(-1.37, -0.51)
resto = r2(t, -0.51, -1.37)

plt.plot(t, resto)

plt.title("Gráfico do resto R2(t)")
plt.xlabel("t")
plt.ylabel("R2(t)")
plt.grid()
plt.show()
```

## Questão 6

Calcule uma estimativa do erro máximo em valor absoluto ao aproximar f(-1.37) por P2(x) em torno de x0.

```python
erro_maximo = np.max(np.abs(resto))
print(erro_maximo)
```

Verifique que o erro máximo é maior que o erro absoluto calculado.
