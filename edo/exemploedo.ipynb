```python
import numpy as np
import matplotlib.pyplot as plt

def g(x, y):
    return np.sqrt(np.abs(y)) - x

x0 = 1
y0 = 9
h = 1
n = 4

def heun(g, x0, y0, h, n):
    x = np.zeros(n + 1)
    y = np.zeros(n + 1)
    x[0] = x0
    y[0] = y0

    for i in range(1, n + 1):
        k1 = h * g(x[i-1], y[i-1])
        k2 = h * g(x[i-1] + h, y[i-1] + k1)
        y[i] = y[i-1] + 0.5 * (k1 + k2)
        x[i] = x[i-1] + h

    return x, y

xHeun, yHeun = heun(g, x0, y0, h, n)
print(yHeun)

plt.plot(xHeun, yHeun, marker='o', label='Heun')
plt.xlabel('x')
plt.ylabel('y')
plt.title('MÃ©todo de Heun')
plt.legend()
plt.grid()
plt.show()
```
